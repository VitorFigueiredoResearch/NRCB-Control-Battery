# NRCB Quickstart: Paper and Figure Reproduction

This notebook is a lightweight quickstart for the **Nisa Radial Control Battery (NRCB)** release package.

It is limited to paper/figure reproduction from compact release artifacts. It:

- reproduces headline summary values from compact tables;
- checks branch sizes and control tiers;
- regenerates manuscript figures in a notebook-local sandbox under `notebooks/_outputs/`;
- demonstrates SPARC validator paper-reproduction modes;
- demonstrates full-production fail-closed behavior when external SPARC/source dependencies are absent.

It does **not** rerun full production, does **not** require raw SPARC data, does **not** download files, and does **not** make physical interpretation or model-validation claims.

## 1. Introduction and scope

The notebook uses only release-local files and relative paths. The compact result tables are already part of the paper-reproduction release. Full production reruns require official external SPARC/source inputs and are intentionally outside this quickstart.

## 2. Locate release root

This cell works when the notebook is launched from either the release root or the `notebooks/` folder.

```python
from pathlib import Path
import csv
import shutil
import subprocess
import sys


def find_release_root(start=None):
    start = Path(start or Path.cwd()).resolve()
    for candidate in (start, *start.parents):
        if (
            (candidate / "results" / "m_surrogate_extension_v1" / "global_control_battery_summary.csv").exists()
            and (candidate / "src" / "validate_sparc_bootstrap.py").exists()
            and (candidate / "figures" / "source" / "generate_control_validation_figures.py").exists()
        ):
            return candidate
    raise RuntimeError("Could not locate NRCB release root from the current working directory.")


RELEASE_ROOT = find_release_root()
NOTEBOOK_DIR = RELEASE_ROOT / "notebooks"
OUTPUT_ROOT = NOTEBOOK_DIR / "_outputs"
FIGURE_OUTPUT_DIR = OUTPUT_ROOT / "figures"
FIGURE_WORKDIR = OUTPUT_ROOT / "figure_sandbox"
RESULT_DIR = RELEASE_ROOT / "results" / "m_surrogate_extension_v1"

print(f"PASS release root: {RELEASE_ROOT}")
print(f"Notebook outputs will be written under: {OUTPUT_ROOT.relative_to(RELEASE_ROOT)}")
```

## 3. Load compact result tables

The notebook reads compact summary tables only. It does not read the large realization table and does not require raw SPARC files.

```python
def load_csv(path):
    with path.open("r", encoding="utf-8-sig", newline="") as handle:
        return list(csv.DictReader(handle))


global_summary = load_csv(RESULT_DIR / "global_control_battery_summary.csv")
per_galaxy_summary = load_csv(RESULT_DIR / "per_galaxy_null_summary.csv")
figure_source = load_csv(RESULT_DIR / "figure2_source_data.csv")

print(f"PASS global rows: {len(global_summary)}")
print(f"PASS per-galaxy rows: {len(per_galaxy_summary)}")
print(f"PASS figure source rows: {len(figure_source)}")
```

## 4. Reproduce headline values

Values are rounded for display. These are descriptive control-distribution summaries from compact release tables, not new computations from raw realizations.

```python
def row_for(rows, branch, tier):
    for row in rows:
        if row["sample_branch"] == branch and row["control_tier"] == tier:
            return row
    raise KeyError((branch, tier))


def as_float(row, column):
    value = row[column]
    if value == "":
        return None
    return float(value)


def approx_check(label, value, expected, tol=5e-4):
    ok = abs(value - expected) <= tol
    status = "PASS" if ok else "FAIL"
    print(f"{status} {label}: {value:.3f} (expected approx {expected:.3f})")
    if not ok:
        raise AssertionError(f"{label}: observed {value}, expected approx {expected}")


partial_abs_shuffle = row_for(global_summary, "partial_rank_134", "shuffle_surrogate_partial_rank_absolute")
partial_abs_shift = row_for(global_summary, "partial_rank_134", "circular_shift_control_partial_rank_absolute")
residual_phase = row_for(global_summary, "spectral_control_76", "random_phase_surrogate_residual")
radial_template_phase = row_for(global_summary, "spectral_control_76", "random_phase_surrogate_radial_template")
exp_ref = row_for(global_summary, "spectral_control_76", "smooth_profile_exponential_reference")
lin_ref = row_for(global_summary, "spectral_control_76", "smooth_profile_linear_reference")
quad_ref = row_for(global_summary, "spectral_control_76", "smooth_profile_quadratic_reference")

approx_check("partial-rank observed absolute median", as_float(partial_abs_shuffle, "observed_median"), 0.509)
approx_check("shuffle absolute control median-of-medians", as_float(partial_abs_shuffle, "control_median_of_medians"), 0.182)
approx_check("exact circular-shift absolute control median-of-medians", as_float(partial_abs_shift, "control_median_of_medians"), 0.532)

res_obs = as_float(residual_phase, "observed_median")
res_ctl = as_float(residual_phase, "control_median_of_medians")
rad_obs = as_float(radial_template_phase, "observed_median")
rad_ctl = as_float(radial_template_phase, "control_median_of_medians")

approx_check("residual random-phase spectral observed", res_obs, 0.676)
approx_check("residual random-phase spectral control", res_ctl, 0.676)
approx_check("radial-template random-phase spectral observed", rad_obs, 0.754)
approx_check("radial-template random-phase spectral control", rad_ctl, 0.754)

approx_check("smooth exponential fixed reference", as_float(exp_ref, "control_median_of_medians"), 0.548)
approx_check("smooth linear fixed reference", as_float(lin_ref, "control_median_of_medians"), 0.553)
approx_check("smooth quadratic fixed reference", as_float(quad_ref, "control_median_of_medians"), 0.570)
```

## 5. Check branch sizes and control tiers

The branch counts are checked from `per_galaxy_null_summary.csv`. The N=134 partial-rank branch and N=76 spectral-control branch remain separated.

```python
branch_galaxies = {}
branch_tiers = {}
for row in per_galaxy_summary:
    branch_galaxies.setdefault(row["sample_branch"], set()).add(row["galaxy_id"])
    branch_tiers.setdefault(row["sample_branch"], set()).add(row["control_tier"])

expected_counts = {"partial_rank_134": 134, "spectral_control_76": 76}
for branch, expected in expected_counts.items():
    observed = len(branch_galaxies.get(branch, set()))
    status = "PASS" if observed == expected else "FAIL"
    print(f"{status} {branch}: {observed} distinct galaxies")
    if observed != expected:
        raise AssertionError(f"{branch}: observed {observed}, expected {expected}")

for branch in sorted(branch_tiers):
    print(f"\n{branch} tiers:")
    for tier in sorted(branch_tiers[branch]):
        print(f"  - {tier}")
```

## 6. Regenerate figures into `notebooks/_outputs/figures/`

The release manuscript figures are fixed artifacts. To avoid modifying them, this cell creates a notebook-local figure sandbox, copies only `figures/source/` and `results/`, runs the figure script inside that sandbox, and copies generated PDF/PNG outputs into `notebooks/_outputs/figures/`.

```python
if FIGURE_WORKDIR.exists():
    shutil.rmtree(FIGURE_WORKDIR)
if FIGURE_OUTPUT_DIR.exists():
    shutil.rmtree(FIGURE_OUTPUT_DIR)

(FIGURE_WORKDIR / "figures").mkdir(parents=True, exist_ok=True)
shutil.copytree(RELEASE_ROOT / "figures" / "source", FIGURE_WORKDIR / "figures" / "source")
shutil.copytree(RELEASE_ROOT / "results", FIGURE_WORKDIR / "results")

script_dir = FIGURE_WORKDIR / "figures" / "source"
cmd = [sys.executable, "generate_control_validation_figures.py"]
run = subprocess.run(cmd, cwd=script_dir, capture_output=True, text=True)
print(run.stdout)
if run.stderr:
    print(run.stderr)
print(f"figure script exit code: {run.returncode}")
if run.returncode != 0:
    raise RuntimeError("Figure regeneration failed inside notebook sandbox.")

FIGURE_OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
generated = sorted((FIGURE_WORKDIR / "figures").glob("*_rev1.*"))
for path in generated:
    if path.suffix.lower() in {".pdf", ".png"}:
        shutil.copy2(path, FIGURE_OUTPUT_DIR / path.name)

pdfs = sorted(FIGURE_OUTPUT_DIR.glob("*.pdf"))
pngs = sorted(FIGURE_OUTPUT_DIR.glob("*.png"))
print(f"PASS generated PDFs: {len(pdfs)}")
print(f"PASS generated PNGs: {len(pngs)}")
for path in pdfs + pngs:
    print(f"  - {path.relative_to(RELEASE_ROOT)}")
if len(pdfs) < 3 or len(pngs) < 3:
    raise AssertionError("Expected at least three PDF and three PNG figure outputs.")
```

## 7. Run SPARC validator manuscript-reproduction mode

Paper/manuscript reproduction does not require raw SPARC files.

```python
def run_validator(mode, extra_args=None, expect_success=True):
    extra_args = extra_args or []
    cmd = [sys.executable, "src/validate_sparc_bootstrap.py", "--mode", mode, "--project-root", ".", *extra_args]
    result = subprocess.run(cmd, cwd=RELEASE_ROOT, capture_output=True, text=True)
    print("$ " + " ".join(cmd))
    print(result.stdout)
    if result.stderr:
        print(result.stderr)
    expected = (result.returncode == 0) if expect_success else (result.returncode != 0)
    status = "PASS" if expected else "FAIL"
    print(f"{status} {mode}: exit code {result.returncode}")
    if not expected:
        raise AssertionError(f"Unexpected validator result for {mode}")
    return result


manuscript_validator = run_validator("manuscript-reproduction", expect_success=True)
```

## 8. Run SPARC validator figure-reproduction mode

Figure reproduction uses compact release results and fixed figure assets. It does not require the large realization table.

```python
figure_validator = run_validator("figure-reproduction", expect_success=True)
```

## 9. Demonstrate full-production fail-closed behavior without external SPARC/source dependencies

This negative check intentionally points full-production mode at a nonexistent source root. A nonzero exit is expected and demonstrates fail-closed behavior. This is not a production run.

```python
missing_source_root = OUTPUT_ROOT / "nonexistent_source_root_for_fail_closed_demo"
if missing_source_root.exists():
    shutil.rmtree(missing_source_root)

negative_full_production = run_validator(
    "full-production",
    extra_args=["--source-root", str(missing_source_root)],
    expect_success=False,
)
print("PASS full-production negative test failed closed as expected.")
```
